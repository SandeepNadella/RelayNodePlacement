```python
%matplotlib inline
import sys

import matplotlib.pyplot as plt
import networkx as nx
import random
#importing graphviz for fixed graph layout. graphviz should be installed using homebrew.
#$brew install graphviz 
#$pip install pygraphviz
from networkx.drawing.nx_agraph import graphviz_layout


#creating the source relay node graph with random edges
N=random.randint(2,20)
G= nx.cycle_graph(N)
for i in range(random.randint(0,20)):
    G.add_edge(random.randint(0,N-1),random.randint(0,N-1))
for (a,b) in G.edges():
    G.edges[a,b]['weight'] = random.randint(0,20)

#creating the source relay node graph with fixed edges
# G = nx.cycle_graph(5)
# G.add_edge(0,1,weight=2)
# G.add_edge(0,3,weight=3)
# G.add_edge(0,4,weight=3)
# G.add_edge(1,2,weight=3)
# G.add_edge(2,3,weight=5)
# G.add_edge(3,4,weight=2)
G.pos = graphviz_layout(G)

#computing the minimum spanning tree(MST)
Tp = nx.minimum_spanning_tree(G)
Tp.pos = graphviz_layout(G)

#ploting the source relay node graph G
plt.figure(figsize=(20,14))
pos=G.pos
edge_labels=nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels, font_size=18, font_color='black', font_family='sans-serif', font_weight='bold')
nx.draw_networkx(G,pos,with_labels=True,node_size=1000, node_color='red', font_size=18, font_color='white', font_family='sans-serif', font_weight='bold')
plt.show()

plt.clf

#plotting the MST Tp
plt.figure(figsize=(20,14))
pos=Tp.pos
edge_labels=nx.get_edge_attributes(Tp,'weight')
nx.draw_networkx_edge_labels(Tp,pos,edge_labels, font_size=18, font_color='black', font_family='sans-serif', font_weight='bold')
nx.draw_networkx(Tp,pos,with_labels=True,node_size=1000, node_color='red', font_size=18, font_color='white', font_family='sans-serif', font_weight='bold')
plt.show()

```
